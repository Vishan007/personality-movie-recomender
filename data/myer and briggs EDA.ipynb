# EDA for the dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('16p.csv' , encoding='latin-1' , index_col = 'Response Id')
```

```python
questions = df.columns[0:60]
```

```python
##changing the names of columns
col = []
i = 1
for que in questions:
    col_name = "Q{}".format(i)
    i = i+1
    col.append(col_name)
```

```python
x = df.drop('Personality',axis=1)
y = df['Personality']
x.columns = col
```

```python
x.describe()
```

```python
x.isnull().sum()
```

### No null values are found

```python
plt.figure(figsize = (10,10)) 
sns.countplot(y)          ##checking if the data is balanced
```

### All the personalities are represented almost equally in the dataset so we can conclued that the data is balanced.

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
y = label.fit_transform(y)
y = pd.DataFrame(y , columns=['personality'])
```

```python
##mergening the dataset to find the corelation matrix
frames = [x ,y]
data = pd.concat(frames,axis=1)
```

```python
data
```

```python
corr = data.corr()
corr
```

```python
plt.figure(figsize = (20,20)) 
sns.heatmap(corr  ,cmap='terrain')
```

## Feature selection

```python
from sklearn.feature_selection import mutual_info_regression
mutual_info = mutual_info_regression(x,y)
```

```python
mutual_info = pd.Series(mutual_info)
mutual_info.index = x.columns
mutual_information_features = mutual_info.sort_values(ascending=False)
```

```python
mutual_information_features
```

```python
mutual_info.index = questions
mutual_information_questions = mutual_info.sort_values(ascending=False)
questions_df=pd.DataFrame(mutual_information_questions )
```

```python
imp_question = []
for i in questions_df.index.values:
    imp_question.append(i)
    if len(imp_question) == 16:
        break
```

```python
X = pd.read_csv('16p.csv',encoding='latin-1',usecols=imp_question)
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score , confusion_matrix , classification_report
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=3)
```

```python
log_model = LogisticRegression()
log_model.fit(X_train,y_train)
y_train_pred=log_model.predict(X_train)
```

```python
print('The train accuracy score is',accuracy_score(y_train_pred,y_train))
```

```python
y_test_pred = log_model.predict(X_test)
```

```python
print('The test accuracy score is',accuracy_score(y_test_pred,y_test))
```

```python
log_con_mat = confusion_matrix(y_test_pred,y_test)
sns.heatmap(log_con_mat,annot=True , cbar=False , fmt='.6g')
```

```python
from sklearn import tree
dtc = tree.DecisionTreeClassifier()
dtc.fit(X_train,y_train)
dtc_train_pred = dtc.predict(X_train)
dtc_test_pred = dtc.predict(X_test)
print('The train accuracy of decision tree model is ',accuracy_score(dtc_train_pred,y_train))
print('The test accuracy of decision tree model is ',accuracy_score(dtc_test_pred,y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train , y_train)
rfc_y_pred = rfc.predict(X_test)
rfc_score = accuracy_score(rfc_y_pred , y_test)
print('The test accuracy of random forest model is ',rfc_score)
```

```python
rfc_con_mat = confusion_matrix(rfc_y_pred,y_test)
sns.heatmap(rfc_con_mat,annot=True , cbar=False , fmt='.6g')
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
rfc.fit(X_train_scaled,y_train)
y_pred_scaled = rfc.predict(X_test)
```

```python
accuracy_score(y_pred_scaled,y_test)
```

```python
from sklearn.svm import SVC
svc_model = SVC()
svc_model.fit(X_train_scaled,y_train)
y_pred_svc = svc_model.predict(X_test_scaled)
print('The accuracy score of svc model ',accuracy_score(y_pred_svc,y_test))
```

```python
svc_con_mat = confusion_matrix(y_pred_svc,y_test)
sns.heatmap(svc_con_mat,annot=True , cbar=False , fmt='.6g')
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train_scaled,y_train)
knn_y_pred = knn.predict(X_test_scaled)
print('The accuracy score of knn',accuracy_score(knn_y_pred,y_test))
```
